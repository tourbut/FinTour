```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
qyld = pd.read_csv('qyld_chart.csv', index_col=0, parse_dates=True)
qyld_div = pd.read_csv('qyld_div.csv', index_col=0, parse_dates=True)
qyld['Dividends'] = qyld_div['Dividends']
qyld = qyld[qyld.index>='2022-01-01'].reset_index()
qyld.tail(10)
```

```python
fig, ax1 = plt.subplots()
ax1 = plt.plot(qyld['Close'], color="blue", label="Close")
ax2 = plt.twinx()
ax2 = plt.plot(qyld['Dividends'], color="red", label="Dividends",marker='o')
```

```python
# Filter data from 2022-01-01
df_qyld_2022 = qyld[qyld['Date'] >= '2022-01-01']

# Find dates with dividends
dividend_dates_2022 = df_qyld_2022[df_qyld_2022['Dividends'].notna()]['Date']

# Plot close prices
plt.figure(figsize=(14, 7))
plt.plot(df_qyld_2022['Date'], df_qyld_2022['Close'])

# Add larger red dots for dividend dates
for date in dividend_dates_2022:
    close_value = df_qyld_2022[df_qyld_2022['Date'] == date]['Close'].values
    if close_value.size > 0:
        plt.scatter(date, close_value[0], color='red', s=50)

plt.title('Close Prices from 2022 with Dividend Dates')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()
```
