```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
qyld = pd.read_csv('qyld_chart.csv', index_col=0, parse_dates=True)
qyld_div = pd.read_csv('qyld_div.csv', index_col=0, parse_dates=True)
qyld['Dividends'] = qyld_div['Dividends']
qyld = qyld[qyld.index>='2022-01-01']
qyld.tail(10)
```

```python
fig, ax1 = plt.subplots()
ax1 = plt.plot(qyld['Close'], color="blue", label="Close")
ax2 = plt.twinx()
ax2 = plt.plot(qyld['Dividends'], color="red", label="Dividends",marker='o')
```
